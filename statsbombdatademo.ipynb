Statsbomb posiada płatną subskrypcję, ale do celów edukacyjnych udostępniają za darmo wybrane sezony/mistrzostwa. W tym przypadku La Liga sezon 15/16.

```python
from statsbombpy import sb
import pandas as pd
```

```python
sb.competitions().head()
```

```python
sb.matches(competition_id=11, season_id=27).head()
```

```python
team = "Barcelona"
all_matches = sb.matches(competition_id=11, season_id=27)
team_matches = all_matches[(all_matches['home_team'] == team) | (all_matches['away_team'] == team)]

goal_data = []
for match_id in team_matches['match_id']:

    events = sb.events(match_id)
    team_goals = events[(events['type'] == 'Shot') & (events['shot_outcome'] == 'Goal') & (events['team'] == team)]
    goal_data.append(team_goals)

goals = pd.concat(goal_data)
```

```python
goals_relevant = goals[['match_id', 'timestamp', 'player', 'shot_type']].reset_index()
goals_relevant
```

```python
goals_messi = goals_relevant[goals_relevant['player'] == "Lionel Andrés Messi Cuccittini"].reset_index()
goals_messi
```
